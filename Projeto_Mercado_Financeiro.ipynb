```python
# Importação de Bibliotecas

import numpy as np
import pandas as pd
pd.__version__

# analise gráfica

import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go

# ignorar avisos

import warnings

# desabilitando avisos
warnings.filterwarnings('ignore')
```

```python
#Lendo o arquivo, acabei utilizando direto do computador, não conseguindo sincronizar com o google drive
df = pd.read_excel('/content/Case_004 - Magalu - Sem Resoluçao.xlsx')
df.head()
```

```python
#verificar a dimensão dos dados
df.shape
```

```python
#verificar o "tipo"
df.info()
```

```python
#verificar média,mediana,moda e quartis
df.describe()
```

```python
# Series Temporais

dados = df.set_index("Data") #colocando a Data como index

dados.head()
```

```python
plt.style.use('seaborn-darkgrid')
plt.figure(figsize=(16,5))
plt.title('Analise Ações Magalu', fontsize=15, loc='left')
plt.plot (dados.index, dados['Fechamento'])

plt.xlabel('Periodo da cotação')
plt.ylabel('Valor da Ação R$');
```

```python
dados.tail() #verificar os 5 últimos
```

```python
media_movel= dados['Fechamento'].rolling(5).mean()
media_tendencia = dados['Fechamento'].rolling(30).mean()

plt.style.use('seaborn-darkgrid')
plt.figure(figsize=(16,5))
plt.title('Analise Ações Magalu', fontsize=15, loc='left')

plt.plot (dados.index, dados['Fechamento'])
plt.plot (dados.index, media_movel)
plt.plot (dados.index, media_tendencia)
plt.xlabel('Periodo da cotação')
plt.ylabel('Valor da Ação R$');
```

```python
sns.boxplot(data=dados, x='Fechamento')
```

```python
#MELHORANDO O BOXPLOT, mensal.
df['Mes'] = df['Data'].dt.month

plt.figure(figsize=(16,5))
sns.boxplot(data=df, x='Mes',y='Fechamento')
```

```python
df.groupby(['Mes']).describe()['Fechamento']
```

```python
#PLOT DE BOVESPA, Gráfico mais profissional

grafico = go.Figure(
    data=[
        go.Candlestick(
            x=dados.index,
            open=dados['Abertura'],
            high=dados['Maior'],
            low=dados['Menor'],
            close=dados['Fechamento']
        )
    ]
)
#grafico.update_layout (xaxis_rangeslider_visible=False) #PARA REMOVER A BARRA ABAIXO DO GRAFICO
grafico.show()
```
